```python


import random
import re
import smtplib
from email.message import EmailMessage

# function to generate 6 digit otp
def generate_otp():
  otp=str(random.randint(100000,999999))
  return otp



# function to start sending the otp to the user's email address using SMTP
# Connect to the SMTP server
def setup_gmail():
  server = smtplib.SMTP('smtp.gmail.com',587)
  server.starttls() # Start TLS for security
  from_mail='adhvikaayansh06@gmail.com'
  server.login(from_mail,'thjt fqsf dmlt kmzv')
  to_mail=input("enter the email_id to send otp: ")

  #Error handing first validate the email address to send the OTP
  if re.match(r"^[\w\.-]+@[\w\.-]+\.\w+$",to_mail):
    print("Entry is valid!")
    # content of the email
    # Create an instance of EmailMessage
    msg=EmailMessage()
    # Set the email subject
    msg['subject']='otp verification mail'
    # Set the sender's email address
    msg['from']=from_mail
    # Set the recipient's email address
    msg['to']=to_mail
    # Set the content of the email
    msg.set_content('Your otp is: '+generated_otp)

    server.send_message(msg)# Send the email
    print('Email sent')
    server.quit()# Disconnect from the server

    #verify otp function
    verify_otp()
  else:
    print(" Email is not valid please try again !")
# created verify_otp function to prompt the user to enter the OTP received in their mail and to verify the entered otp matches the generated otp
def verify_otp():
  for attempt in range(0,3):
    input_otp=input('Enter the otp: ')
    #Error handling: verify the otp matches
    if input_otp==generated_otp:
      print("OTP verified successfully!Access granted!")
      break
    #Allow the user user to enter the otp three times if wrong otp enterd
    else:
      print('Invalid OTP,Try Again!This is attempt',attempt+1)
      attempt+=1
      if attempt==3:
        print("Access denied!Maximum attempt reached!")
        break
generated_otp=generate_otp()
print(generated_otp)
setup_gmail()
print('Thank you for using OTP vrification system')
```
